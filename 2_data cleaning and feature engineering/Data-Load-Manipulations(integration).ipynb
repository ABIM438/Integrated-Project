```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

insurance_df = pd.read_excel('C:/github/Integration_Project/datasets/insurance_claims_raw.xlsx')
insurance_df.head()
```

```python
missing_values = insurance_df.isnull().sum()
print(missing_values)
```

```python
numeric_means = insurance_df.select_dtypes(include=[np.number]).mean()
insurance_df_filled = insurance_df.fillna(numeric_means)
categorical_mode = insurance_df.select_dtypes(include=[object]).mode().iloc[0]
insurance_df_filled.update(insurance_df.select_dtypes(include=[object]).fillna(categorical_mode))
```

```python
missing_values_new = insurance_df_filled.isnull().sum()
print(missing_values_new)
```

```python
insurance_df_filled.drop(columns=["_c39"], inplace=True)
```

```python
insurance_df_filled.to_excel('C:/github/Integration_Project/datasets/insurance_claims_processed.xlsx', index=False)
```
